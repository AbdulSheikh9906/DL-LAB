```

# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```

# Automatically Download Dataset

dataset_url = "https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/mushrooms.csv"

df = pd.read_csv(dataset_url)

print("Dataset Loaded Successfully!")
print("Dataset Shape:", df.shape)

df.head()
```

```

# Dataset Information

print(df.info())

print("\nClass Distribution:\n")
print(df['type'].value_counts())
```

```

# Rename Target Column

df.rename(columns={'type':'class'}, inplace=True)

print(df.columns)
```

```

# Encode Categorical Features

from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

for column in df.columns:
    df[column] = encoder.fit_transform(df[column])

print("Encoding Completed Successfully!")
df.head()
```

```

# Split Dataset

from sklearn.model_selection import train_test_split

X = df.drop('class', axis=1)
y = df['class']

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42
)

print("Training Shape:", X_train.shape)
print("Testing Shape:", X_test.shape)
```

```

# Train Naive Bayes Model

from sklearn.naive_bayes import GaussianNB

model = GaussianNB()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Model Training Completed!")
```

```

# Accuracy Score

from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)
```

```

# Confusion Matrix

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

print("Confusion Matrix:\n")
print(cm)

plt.figure(figsize=(6,4))

sns.heatmap(cm, annot=True, fmt='d')

plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.show()
```


# Conclusion

The Naive Bayes classifier was successfully implemented on the Mushroom dataset.
The model achieved high classification accuracy for edible and poisonous mushrooms.
