- Automatically downloads dataset from Kaggle

- Visualizes lower dimensional data

```

# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import kagglehub
import os

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```

# Download Dataset from Kaggle

path = kagglehub.dataset_download("tomigelo/spotify-audio-features")

print("Dataset downloaded successfully!")
print("Dataset path:", path)
```

```

# Load CSV File

csv_file = os.path.join(path, "SpotifyAudioFeaturesNov2018.csv")

df = pd.read_csv(csv_file)

print("Dataset Loaded Successfully!")
print("Shape of dataset:", df.shape)

df.head()
```

```

# Dataset Information

df.info()
```

```

# Select Only Numerical Columns

numerical_df = df.select_dtypes(include=np.number)

print("Numerical Dataset Shape:", numerical_df.shape)

numerical_df.head()
```

```

# Data Standardization

scaler = StandardScaler()

scaled_data = scaler.fit_transform(numerical_df)

print("Data Standardized Successfully!")
```

```

# Apply PCA

pca = PCA()

pca.fit(scaled_data)

print("PCA Applied Successfully!")
```

```

# Scree Plot

plt.figure(figsize=(10,6))

plt.plot(
    range(1, len(pca.explained_variance_ratio_) + 1),
    pca.explained_variance_ratio_,
    marker='o'
)

plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot')
plt.grid(True)

plt.show()
```

```

# PCA with 2 Components

pca_2 = PCA(n_components=2)

pca_data = pca_2.fit_transform(scaled_data)

print("Reduced Data Shape:", pca_data.shape)
```

```

# Create Lower Dimensional DataFrame

pca_df = pd.DataFrame(
    data=pca_data,
    columns=['PC1', 'PC2']
)

pca_df.head()
```

```

# 2D Visualization

plt.figure(figsize=(10,7))

plt.scatter(
    pca_df['PC1'],
    pca_df['PC2'],
    alpha=0.5
)

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA - 2D Visualization of Spotify Audio Features')
plt.grid(True)

plt.show()
```

```

# Explained Variance Ratio

explained_variance = pca_2.explained_variance_ratio_

print("Explained Variance Ratio:")
print(explained_variance)
```
